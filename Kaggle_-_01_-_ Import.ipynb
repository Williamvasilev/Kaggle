# 01 - Import

## Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.display import display, Markdown
sns.set_style("darkgrid")
pd.set_option('display.max_columns', None)  
from sklearn.model_selection import train_test_split

import sys, os, yaml

SEED = 12312
```

## Dataset

```python
df = pd.read_csv("orig/train.csv")
print(df.shape)
df.head()
```

```python
df.dtypes.value_counts()
```

```python
df.Target.value_counts(dropna=False)
```

```python
print(df.shape)
df = df.dropna(subset = ['Target'])
print(df.shape)
```

```python
df.isna().sum().value_counts()
```

```python
print(df.shape)
df = df.dropna()
print(df.shape)
```

```python
df_train, df_test = train_test_split(df, stratify=df['Target'], test_size = 0.40, random_state=SEED)
df_train.shape, df_test.shape
```

```python
df_score = pd.read_csv("orig/test.csv")
print(df_score.shape)
df_score.head()
```

## Output

```python
df_train.to_pickle("data/train.pkl")
df_test.to_pickle("data/test.pkl")
df_score.to_pickle("data/score.pkl")
```
